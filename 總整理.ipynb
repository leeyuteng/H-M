```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_test = pd.read_csv('test(include_y)_data_label0417.csv',dtype={'article_id': str})
df_train = pd.read_csv('train(include_y)_data_label0417.csv',dtype={'article_id': str})
```

```python
df_test
```

```python
df_train
```

```python
X_test = df_test[['price','cluster_age_avgconsume','alive','Buy_n_year',\
                 'loyaty','avg_consume','age_class','attribute','graphical_rename',\
                 'color_rename','garment_group_name']]
y_test = df_test['Buy']
X_train = df_train[['price','cluster_age_avgconsume','alive','Buy_n_year',\
                 'loyaty','avg_consume','age_class','attribute','graphical_rename',\
                 'color_rename','garment_group_name']]
y_train = df_train['Buy']
```

```python
X_test.info()
```

```python
X_train
```

```python
with open ('xgboost.pickle','rb') as fr:
    xgbc = pickle.load(fr)
```

```python
XGB_pred=xgbc.predict(X_test)
print(XGB_pred)
```

```python
XGBtrain_score=xgbc.score(X_train, y_train)
print(XGBtrain_score)
```

```python
XGBtest_score=xgbc.score(X_test,y_test)
print('The Accuracy of XGBOOST Classifier on testing set:', XGBtest_score)
```

```python
# Sckit-Learn
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,XGB_pred))
print(confusion_matrix(y_test,XGB_pred))
```

```python
with open ('LGBM.pickle','rb') as fr:
    lgbmc = pickle.load(fr)
```

```python
LGBM_pred=lgbmc.predict(X_test)
print(LGBM_pred)
```

```python
LGBMtrain_score=lgbmc.score(X_train, y_train)
print(LGBMtrain_score)
```

```python
LGBMtest_score=lgbmc.score(X_test,y_test)
print('The Accuracy of LightGBM Classifier on testing set:', LGBMtest_score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,LGBM_pred))
print(confusion_matrix(y_test,LGBM_pred))
```
