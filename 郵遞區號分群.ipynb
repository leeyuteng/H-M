```python
import pandas as pd
import numpy as np
import pickle
```

```python
cus = pd.read_csv("cus.csv")
cus
```

```python
postal = cus['postal_code'].value_counts()
postal = pd.DataFrame(postal)
postal.reset_index(inplace=True)
postal
```

```python
city1 = postal[:1]
city1['city'] = 0
city1
```

```python
city2 = postal[1:126]
city2['city'] = 1
city2
```

```python
city3 = postal[126:]
city3['city'] = 2
city3
```

```python
#上下合併 concat
#df=pd.concat([df,df2],axis=0)
city = pd.concat([city1,city2],ignore_index=True)
city = pd.concat([city,city3],ignore_index=True)
city
```

```python
city.drop('postal_code',axis=1,inplace=True)
```

```python
city.rename(columns={'index': 'postal_code'}, inplace=True)
city
```

```python
city
```

```python
cus_city = pd.merge(cus,city,on='postal_code',how="inner")
cus_city
```

```python
cus_city['city'].value_counts()
```

```python
cus_city
```

```python
cus_city.to_csv('cus最新_city.csv',index=True)
```
